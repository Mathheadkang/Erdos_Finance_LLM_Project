```python
import yfinance as yf      
import pandas as pd   
def get_news(stock:str) -> list:
    
    try:
        ticker = yf.Ticker(stock)
        news = ticker.news 
        
        if not news:
            print(f'No news found for {stock}')
            return [] 
        
        #relevant_news = [
            #item for item in news if item.get('content',{}).get('contentType')
            #== 'STORY'
        #]
        
        all_news = [] 
        #for i, item in enumerate(relevant_news):
        for i, item in enumerate(news):
            
            try:
                content = item.get('content', {})
                current_news = {
                    'title': content.get('title'), 
                    'summary': content.get('summary'), 
                    'url' : content.get('canonicalUrl', {}).get('url'), 
                    'pubdate': content.get('pubDate', '').split('T')[0],
                }
                all_news.append(current_news)
            except Exception as e:
                print(f'Error processing news item {i}: {e}')
                continue   
        return all_news
    except Exception as e:
        print(f'Error fetching news for {stock}: {e}')
        return []
            
            
```

```python
from markitdown import MarkItDown 
import requests
import re

session = requests.Session()
session.headers.update({
    'User-Agent': 'python-requests/2.32.3', 
    'Accept-Encoding': 'gzip, deflate', 
    'Accept':'*/*', 
    'Connection': 'keep-alive'
})
md = MarkItDown(requests_session=session)
```

```python
def remove_links(text):
    if text is None:
        return ''
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'[#*()+\-\n]','', text)
    text = re.sub(r'/\S*', '', text)
    text = re.sub(r'  ', '',text)
    
    return text
    
```

```python
def extract_news(link):
    extracted = md.convert(link)
    title = extracted.title.strip() 
    content = extracted.text_content.strip() 
    return title, "\n" + remove_links(content)
```

```python
def extract_full_news(stock: str) -> list:
    news_list = get_news(stock)
    for i, item in enumerate(news_list):
        try:
            full_text = extract_news(item['url'])
            item['full_news'] = full_text 
        except Exception as e:
            print(f'Error extracting full news for {stock} item {i}: {e}')
            continue
    return news_list
```

```python
NEWS_ANALYSIS_PROMPT = """You are a specialized financial news analyzer. Your task is to analyze stock news articles and provide structured output.

Given news articles about a specific stock, you must:
1. Determine sentiment for each article (Positive, Negative, or Neutral)
2. Extract key points relevant to the stock price
3. Provide an overall summary of the news
4. Make an investment recommendation based on the news

IMPORTANT: Return ONLY a properly formatted JSON with no extra text, explanations, or conversation. The JSON must have this exact structure:
{
    "sentiment_analysis": {
        "title1": "Positive/Negative/Neutral",
        "title2": "Positive/Negative/Neutral",
        ...
    },
    "key_points": [
        "key point 1",
        "key point 2",
        ...
    ],
    "overall_summary": "concise summary of all articles",
    "investment_recommendation": "buy/sell/hold recommendation with brief rationale"
}
"""
```

```python



def analyze_news_with_gpt(stock:str):
    news_data = extract_full_news(stock)
    
    articles_text = "\n\n".join([
        f"Title: {n['title']}\n Content: {n['summary']}"for n in news_data if'summary'in n
    ])
    system_msg = SystemMessage(content=NEWS_ANALYSIS_PROMPT)
    user_msg = HumanMessage(content=f"Stock: {stock}\n News:\n{articles_text}")
    
    response = chat.invoke([system_msg, user_msg])
    return response.content 
    
```
